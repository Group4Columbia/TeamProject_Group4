```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
covid_19_df = pd.read_csv('Project Resources/covid_map_data.csv')
covid_19_df.head(5)
```

```python
final_df = covid_19_df.drop(columns=["people_fully_vaccinated", "total_vaccinations_per_hundred", "people_vaccinated_per_hundred", "people_fully_vaccinated_per_hundred", "daily_vaccinations_per_million", "Note", "new_cases", "new_deaths", "handwashing_facilities", "hospital_beds_per_thousand", "Country", "Status", "diabetes_prevalence", "life_expectancy"])
```

```python
covid_final_df = final_df.dropna()
covid_final_df.head()
```

```python
covid_final_df["total_cases"] = covid_final_df["total_cases"].astype(int)
covid_final_df.dtypes
```

```python
covid_final_df = covid_final_df[covid_final_df.human_development_index >=.7]
covid_final_df.head()
```

```python
covid_final_df = covid_final_df[covid_final_df.human_development_index <=.9]
covid_final_df.head()
```

```python
### Relationship Betweeen HDI and Total Cases
plt.scatter(covid_final_df.human_development_index, covid_final_df.total_cases)
plt.xlabel("Human Development Index")
plt.ylabel("Global Total Cases")
plt.show()
```

```python
X = covid_final_df.human_development_index.values.reshape(-1,1)
```

```python
y = covid_final_df.total_cases
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
### HDI Range of .7-.9
plt.scatter(X, y)
plt.plot(X, y_pred, color='green')
plt.xlabel("Human Development Index")
plt.ylabel("Global Total Cases")
plt.title("COVID-19 Total Cases by Human Development Index")
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
hdi_column = covid_final_df['human_development_index']
cases_column = covid_final_df['total_cases']
```

```python
correlation = hdi_column.corr(cases_column)
correlation
```

# This negative relationship between countries with a HDI between .7 and .9 and total cases was to be expected. As a countries Human Development Index, or HDI, gets better, the less cases there are going to be. The correlation between the two variables is -36.79%. Countries with a greater HDI probably have more access to healthcare facilities and healthcare professionals, which in turn will slow down the spread of cases of COVID-19.
