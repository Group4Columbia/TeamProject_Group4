# The Correlation between the Human Development Index and Total Vaccinations becomes 15.40% a little higher as we analyze countries with a HDI greater than .

```python
covid_final_df = covid_final_df[covid_final_df.human_development_index <=.9]
covid_final_df.head()
```

```python
### Relationship Betweeen HDI and Vaccinations for countries with an HDI between .7 and .9
plt.scatter(covid_final_df.human_development_index, covid_final_df.total_vaccinations_)
plt.xlabel("Human Development Index")
plt.ylabel("Global Total Vaccinations")
plt.title("COVID-19 Vaccinations by Human Development Index")
plt.show()
```

```python
X = covid_final_df.human_development_index.values.reshape(-1,1)
```

```python
y = covid_final_df.total_vaccinations_
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='green')
plt.xlabel("Human Development Index")
plt.ylabel("Global Total Vaccinations")
plt.title("COVID-19 Vaccinations by Human Development Index")
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```
